> ### Signals are **Real** Valued or **Complex** Valued
> * Real Valued Signals ($x: \mathbb R \rightarrow \mathbb R$)
>   * Continuous: Consider signal $325\sin(2\pi \omega_0 t)$.

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
plt.rcParams['text.usetex'] = True

t = np.linspace(-0.02, 0.05, 1000)
omega0 = 100
plt.plot(t, 325 * np.sin(2*np.pi*omega0*t))
plt.xlabel('$t(s)$')
plt.ylabel('$x(t)$')
plt.title(f'$325\sin(2\pi{omega0}t)$')
```

> * Real Valued Signals ($x: \mathbb R \rightarrow \mathbb R$)
>   * Discrete: Consider signal $325\sin(2\pi \omega_0 n \Delta t)$.

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
plt.rcParams['text.usetex'] = True

n = np.arange(50)
dt = 0.07/50
omega0 = 50
plt.stem(n, 325 * np.sin(2*np.pi*omega0*n*dt))
plt.xlabel('$n$')
plt.ylabel('$x[n]$')
plt.title(f'$325\sin(2\pi{omega0}n\Delta t)$')
```

> * Complex Valued Signals ($x: \mathbb R \rightarrow \mathbb C$)
>   * Continuous: Consider signal $x(t) = e^{j \omega_0 t}$.

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
plt.rcParams['text.usetex'] = True

t = np.linspace(-0.02, 0.05, 1000)
omega0 = 100
plt.figure(figsize=(10, 5))
plt.subplot(2, 2, 1)
plt.plot(t, np.exp(2j*np.pi*omega0*t).real)
plt.xlabel('$t(s)$')
plt.ylabel('$\Re(x(t))$')
plt.title('Real part of $x(t) = e^{j100 \pi t}$')
plt.subplot(2, 2, 2)
plt.plot(t, np.exp(2j*np.pi*omega0*t).imag)
plt.xlabel('$t(s)$')
plt.ylabel('$\Im(x(t))$')
plt.title('Imaginary part of $x(t) = e^{j100 \pi t}$')

plt.subplot(2, 2, 3)
plt.plot(t, np.abs(np.exp(2j*np.pi*omega0*t)))
plt.xlabel('$t(s)$')
plt.ylabel('$|x(t)|$')
plt.title('Absolute value of $x(t) = e^{j100 \pi t}$')
plt.subplot(2, 2, 4)
phase = np.angle(np.exp(2j*np.pi*omega0*t))*360/(2*np.pi)
plt.plot(t, phase)
plt.xlabel('$t(s)$')
plt.ylabel('$\\angle x(t) $')
plt.title('Phase of $x(t) = e^{j100 \pi t}$')
phase_max = max(phase)
phase_min = min(phase)
print(phase_max)
plt.plot(t, np.tile(phase_max, np.shape(t)), "r--")
plt.plot(t, np.tile(phase_min, np.shape(t)), "r--")
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
```

# Fourier Series
<p align="center">
    <img src = 'Fig1.png' alt="drawing" width="400"/>
</p>

$$
y(t) = x(t) * h(t) = \int\limits_{ - \infty }^\infty  {h(\tau )x(t - \tau )d\tau  = \int\limits_{ - \infty }^\infty  {h(\tau ){e^{s(t - \tau )}}d\tau  = {e^{st}}\int\limits_{ - \infty }^\infty  {h(\tau ){e^{ - s\tau }}d\tau  = {e^{st}}H(s)} } }
$$
We call $x(t)=e^{st}$ as a **Special Function of LTI Systems**. So if we could write any input to our system as a $x(t) = {a_1}{e^{{s_1}t}} + {a_2}{e^{{s_2}t}} +  \ldots $ then the output could easily calculate as:
$$
y(t) = {a_1}H({s_1}){e^{{s_1}t}} + {a_2}H({s_2}){e^{{s_2}t}} +  \ldots
$$
If we consider $s = j \omega$ then our analysis in the **Fourier** domain. any continuous and periodic signal where implies Dirichlet conditions could write as a sum of exponential functions:
$$
x(t) =  \ldots  + {a_{ - 2}}{e^{ - 2j{\omega _0}t}} + {a_{ - 1}}{e^{ - j{\omega _0}t}} + {a_0} + {a_1}{e^{j{\omega _0}t}} + {a_2}{e^{2j{\omega _0}t}} +  \ldots
$$
> - ### Continues
>$$
> x(t) = \sum\limits_{k =  - \infty }^\infty  {{a_k}{e^{jk{\omega _0}t}}}
>$$
>$$
> {a_k} = \frac{1}{T}\int_T {x(t){e^{ - jk{\omega _0}t}}dt} ,{a_0} = \frac{1}{T}\int_T {x(t)dt}
>$$
> - ### Discrete
>$$
> x[n] = \sum\limits_{k =  < N > } {{a_k}{e^{jk{\omega _0}n}}}
>$$
>$$
>{a_k} = \frac{1}{N}\sum\limits_{n =  < N > } {x[n]{e^{ - jk{\omega _0}n}}} ,{a_0} = \frac{1}{N}\sum\limits_{n =  < N > } {x[n]}
>$$

```python
import sympy
import numpy as np
import matplotlib.pylab as plt
%config InlineBackend.figure_format = 'svg'
plt.rcParams['text.usetex'] = True

t = sympy.Symbol('t',real = True)
T0 = sympy.Symbol('T0', real = True, positive = True)
k = sympy.Symbol('k', integer = True)
omega0 = 2*sympy.pi / T0
x = sympy.exp(-t);x
```

Compute the Fourier transform analytically:

```python
ak = sympy.integrate(1/T0 * x * sympy.exp(-sympy.I * k * omega0 * t), (t, 0, T0))
```

```python
ak
```

```python
ak = ak.simplify(); ak
```

```python
ak.subs(T0, 1)
```

```python
abs_ak = sympy.Abs(ak).simplify(); abs_ak
```

Convert to a numerical pyhton function so the values can be extracted numerically and plotted:

```python
ak_as_numerical = sympy.lambdify([k, T0], ak)
ks = np.arange(-20, 20, 1)
fig, ax = plt.subplots(1, 2, figsize= [10, 4])
ax[0].stem(ks, np.abs(ak_as_numerical(k = ks, T0 = 4)))
ax[0].set_xlabel('$k$')
ax[0].set_ylabel('$|a_k|$')

ax[1].stem(ks, np.angle(ak_as_numerical(k = ks, T0 = 4)) * (180/np.pi))
ax[1].set_xlabel('$k$')
ax[1].set_ylabel('$\\angle(a_k)$')

max(np.angle(ak_as_numerical(k = ks, T0 = 4)) * (180/np.pi))
```

> If $x(t)$ is **periodic real valued** signal defined over the interval $(a, b)$:
> $$x(t) = \frac{a_0}{2} + \sum_{n=1}^{\infty}(a_n \cos(\frac{n \pi t}{L}) + b_n \sin(\frac{n \pi t}{L}))$$
> where $L = \frac{b-a}{2}$ is half of the period. so:
> $$ a_0 = \frac{1}{L} \int_{a}^{b}{x(t) dt}$$
> $$ a_n = \frac{1}{L} \int_{a}^{b}{x(t) \cos(\frac{n \pi t}{L}) dt}$$
> $$ b_n = \frac{1}{L} \int_{a}^{b}{x(t) \sin(\frac{n \pi t}{L}) dt}$$


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
%config InlineBackend.figure_format = 'svg'
plt.rcParams['text.usetex'] = True

dx = 0.001
L = np.pi
x = L * np.arange(-1 + dx, 1 + dx, dx)
n = len(x)
nquart = int(np.floor(n/4))

#define hat function
f = np.zeros_like(x)
f[nquart: 2*nquart] = (4/n) * np.arange(1, nquart+1)
f[2*nquart: 3*nquart] = np.ones(nquart) - (4/n)*np.arange(0, nquart)

fig, ax = plt.subplots(1,2, figsize=(15, 4))
ax[0].plot(x, f, '-', color = 'k', LineWidth = 2, label = 'Hat Function')
ax[0].legend()
ax[1].plot(x, f, '-', color = 'k', LineWidth = 2)

# Compute Fourier Series
name = 'Accent'
cmap = get_cmap('tab10')
colors = cmap.colors
ax[1].set_prop_cycle(color = colors)

A0 = np.sum(f * np.ones_like(x)) * dx
fFS = A0/ 2

A = np.zeros(20)
B = np.zeros(20)
for k in range(20):
    A[k] = np.sum(f * np.cos(np.pi*(k+1)*x/L)) * dx
    B[k] = np.sum(f * np.sin(np.pi*(k+1)*x/L)) * dx
    fFS = fFS + A[k] * np.cos(np.pi*(k+1)*x/L) + B[k] * np.sin(np.pi*(k+1)*x/L)
    ax[1].plot(x, fFS, '-')
ax[1].set_xlabel('$time$')
```

* #### Reconstruction Error via Fourier Series Coefficients

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
%config InlineBackend.figure_format = 'svg'
plt.rcParams['text.usetex'] = True

dx = 0.001
L = np.pi
x = L * np.arange(-1 + dx, 1 + dx, dx)
n = len(x)
nquart = int(np.floor(n/4))

#define hat function
f = np.zeros_like(x)
f[nquart: 2*nquart] = (4/n) * np.arange(1, nquart+1)
f[2*nquart: 3*nquart] = np.ones(nquart) - (4/n)*np.arange(0, nquart)


A0 = np.sum(f * np.ones_like(x)) * dx
fFS = (A0/2) * np.ones_like(f)
kmax = 100
A = np.zeros(kmax)
B = np.zeros(kmax)
ERR = np.zeros(kmax)

A[0] = A0/2
ERR[0] = np.linalg.norm(f-fFS) / np.linalg.norm(f)

for k in range(1, kmax):
    A[k] = np.sum(f*np.cos(np.pi*k*x/L)) * dx
    B[k] = np.sum(f * np.sin(np.pi*k*x/L)) * dx
    fFS = fFS + A[k] * np.cos(np.pi*k*x/L) + B[k] * np.sin(np.pi*k*x/L)
    ERR[k] = np.linalg.norm(f-fFS) / np.linalg.norm(f)
    
#thresh = np.median(ERR) * np.sqrt(kmax) * (4/np.sqrt(3))
#r = np.max(np.where(ERR > thresh))

fig, ax = plt.subplots(1,2, figsize=(15, 4))
ax[0].semilogy(np.arange(kmax), A, 'o', label = '$a_k$', color='b',lw =2)
ax[0].semilogy(np.arange(kmax), B, 'o', label = '$b_k$', color='r',lw =2)
#ax[0].semilogy(r, A[r], '*', color = 'green', MarkerSize = 20)
ax[0].set_xlabel('$\#$ of Coefficients')
ax[0].set_ylabel('Amplitude of Coefficients')
ax[0].legend()

ax[1].semilogy(np.arange(kmax), ERR, 'o', color='b',lw =2)
#ax[1].semilogy(r, A[r], '*', color = 'green', MarkerSize = 10)
ax[1].set_xlabel('$\#$ of Coefficients')
ax[1].set_ylabel('Reconstruction Error')
```
